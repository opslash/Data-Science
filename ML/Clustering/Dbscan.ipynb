```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
```

```python
data = pd.read_csv('Mall_Customers.csv')
data
```

```python
data.drop(columns=['CustomerID','Genre','Age'],inplace=True)
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,:2].values
```

```python
from sklearn.cluster import DBSCAN

dbs = DBSCAN(eps=2, min_samples=5)

y_dbs = dbs.fit_predict(X)
```

```python
data['Cluster'] = pd.DataFrame(y_dbs)
data
```
