```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel('Fabric data.xlsx')
df1.head()
```

```python
df1['Fabric_length'].skew()
```

```python
# OR 

sns.histplot(df1['Fabric_length'])
```

```python
# OR

sns.kdeplot(df1['Fabric_length'])
```

```python
# OR SHAPIRO TEST

from scipy import stats
stats.shapiro(df1['Fabric_length'])
```

```python
from statsmodels.stats.weightstats import ztest

ztest = ztest(df1['Fabric_length'],value=150,alternative='two-sided',ddof=0)
ztest
```

```python
df2 = pd.read_excel('Bolt diameter.xlsx')
df2.head()
```

```python
df2['Diameter'].skew()
```

```python
ttest = stats.ttest_1samp(df2['Diameter'],popmean = 10,alternative='two-sided')
ttest
```

```python
df3 = pd.read_excel('Promotion.xlsx')
df3.head()
```

```python
df3['InterestRateWaiver'].skew(),df3['StandardPromotion'].skew()
```

```python
anova_test = stats.f_oneway(df3['InterestRateWaiver'],df3['StandardPromotion'])
anova_test
```

```python
# Proportion test

binom_test = stats.binomtest(496,2000,0.9)
binom_test
```

```python
df5 = pd.read_excel('JohnyTalkers.xlsx')
df5.head()
```

```python
count = pd.crosstab(df5['Person'],df5['Drinks'],margins = True)
count
```

```python
chisquaretest = stats.chi2_contingency(count)
chisquaretest
```
