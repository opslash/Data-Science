```python
# 6.How can you visualize the proportion of students in each region by using pie chart?
region_counts = df['Region'].value_counts().sort_values(ascending=False)
print(region_counts)
plt.pie(region_counts , labels = region_counts.index , autopct='%1.1f%%')
plt.title("Proportion of Students in Each Region")
plt.show()
```

```python
# 7.How can you create a bar plot showing the total number of students in each region?
plt.bar(region_counts.index,region_counts.values)
```

```python
# 8.How can you visualize the average Mark3 for each region by using bar plot?
region_avg_mark3 = df.groupby('Region')['Mark3'].mean()
plt.bar(region_avg_mark3.index,region_avg_mark3.values)
```

```python
# 9.How can you visualize the total attendance by semester and region by using stacked bar plot?
semester_region_attendance = df.groupby(['Semester' , 'Region'])['Attendance'].sum().unstack()
# Create the stacked bar plot
semester_region_attendance.plot(kind='bar', stacked=True, colormap='Set2')
plt.xlabel("Semester")
plt.ylabel("Total Attendance")
plt.title("Total Attendance by Semester and Region")
plt.legend(title="Region")
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.tight_layout()
plt.show()
```

```python
# 10.How can you visualize attendance per semester in an unstacked bar plot by using unstacked bar plot?
semester_attendance_counts = df.groupby('Semester')['Attendance'].count()
# Create the unstacked bar plot
plt.bar(semester_attendance_counts.index, semester_attendance_counts.values)
plt.xlabel("Semester")
plt.ylabel("Attendance Count")
plt.title("Attendance per Semester (Unstacked)")
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.tight_layout()
plt.show()
```

```python
# 11.How can you create subplots to visualize the distribution of Mark1 and Mark2?

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5)) 

# Distribution of Mark1
ax1.hist(df['Mark1'])
ax1.set_xlabel('Mark1')
ax1.set_ylabel('Number of Students')
ax1.set_title('Distribution of Mark1')

# Distribution of Mark2
ax2.hist(df['Mark2'])
ax2.set_xlabel('Mark2')
ax2.set_ylabel('Number of Students')
ax2.set_title('Distribution of Mark2')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# 12.How can you visualize the relationship between Mark1 and Mark2 along with their distributions by using join plot?
import pandas as pd
import matplotlib.pyplot as plt


# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))

# Scatter plot for the relationship
ax.scatter(df['Mark1'], df['Mark2'], alpha=0.7)

# Histograms on edges for distributions
ax.hist(df['Mark1'], bins=10, density=True, histtype='step', color='red', alpha=0.5, linewidth=2, label='Mark1 Distribution')
ax.hist(df['Mark2'], bins=10, density=True, histtype='step', color='blue', alpha=0.5, linewidth=2, label='Mark2 Distribution')

# Labels and title
ax.set_xlabel('Mark1')
ax.set_ylabel('Mark2')
ax.set_title('Relationship between Mark1 and Mark2 with Distributions')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# 13.How can you visualize the distribution of Mark3 by region using violin plot?

# Create the violin plot
sns.violinplot(x='Region', y='Mark3', data=df)

# Set the title of the plot
plt.title('Distribution of Mark3 by Region')

# Show the plot
plt.show()
```

```python
# 14.How can you visualize relationships among Mark1, Mark2, and Mark3 by using pair plot?

# Create the pair plot
sns.pairplot(df[['Mark1', 'Mark2', 'Mark3']])

# Customize the plot (optional)
plt.suptitle('Pair Plot of Mark1, Mark2, and Mark3', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 15.How can you visualize the correlation between Mark1, Mark2, and Mark3 by using Heatmap?

# Calculate the correlation matrix
correlation_matrix = df[['Mark1', 'Mark2', 'Mark3']].corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True)  # Add annotations for cell values

# Customize the plot (optional)
plt.title('Correlation Heatmap of Mark1, Mark2, and Mark3')
plt.show()
```
