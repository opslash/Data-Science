```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.info()
```

```python
df['Exited'].value_counts()
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df['Geography'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
df.isnull().sum()
```

```python
X = pd.get_dummies(df.drop('Exited',axis=1),drop_first=True)
y = df['Exited']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann = Sequential()
ann.add(Dense(input_dim = 11,units = 21, kernel_initializer = 'uniform', activation = 'relu'))
ann.add(Dense(units= 21 , kernel_initializer = 'uniform' , activation = 'relu'))
ann.add(Dense(units= 21 , kernel_initializer = 'uniform' , activation = 'relu'))
ann.add(Dense(units= 1 , kernel_initializer = 'uniform' , activation = 'sigmoid'))
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size = 32, epochs = 100)
```

```python
y_pred_train =  ann.predict(X_train)
y_pred_train = (y_pred_train >= 0.5)

y_pred_test =  ann.predict(X_test)
y_pred_test = (y_pred_test >= 0.5)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

print('Test accuracy :',accuracy_score(y_test,y_pred_test))
print('Train accuracy :',accuracy_score(y_train,y_pred_train))
```

```python
confusion_matrix(y_test,y_pred_test)
```

```python
def build_cross_classifier():
    classifier = Sequential()
    classifier.add(Dense(input_dim = 11, units = 21, kernel_initializer='uniform', activation = 'relu'))
    classifier.add(Dense(units = 21, kernel_initializer='uniform', activation = 'relu'))
    classifier.add(Dense(units = 21, kernel_initializer='uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer='uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

from scikeras.wrappers import KerasClassifier
classfier = KerasClassifier(build_cross_classifier(),batch_size = 32, epochs = 100)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(classfier,X_train,y_train,cv=5,scoring='accuracy').mean()
```

```python
estimator = KerasClassifier(build_cross_classifier())
param_grid = {'batch_size':[16,32], 'epochs':[50,100,150]}
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator,param_grid,scoring='accuracy',cv=5)
grid_result = grid.fit(X_train,y_train)
```
